```python
import numpy as np
lst = [1, 2, 3, 4, 5] # list of integers
arr = np.array(lst)
print(arr)
```

```python
mat = np.array([[1, 2, 3], [4, 5, 6]]) # 2D array (matrix)
print(mat)
print(mat.shape)
print(mat.ndim)
print(mat.size)
print(mat.dtype)
```

```python
a=np.array([1, 2, 3, 4, 5])
b=np.array([10, 20, 30, 40, 50])
print(a + b)          # Element-wise addition
print(a - b)          # Element-wise subtraction
print(a * b)          # Element-wise multiplication
print(a / b)          # Element-wise division
print(a ** 2)        # Element-wise exponentiation
print(np.sqrt(a))    # Element-wise square root
print(np.log(a))     # Element-wise natural logarithm
print(np.sin(a))     # Element-wise sine
```

```python
#matrix multiplication example
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])
C = np.dot(A, B)
print(C)  
print(A.dot(B))
```

```python
#matrix operations
print(np.transpose(A))  # Transpose of matrix A
a=np.array([1,2,3,4,5,6])
print(np.mean(a))
print(np.median(a))
print(np.std(a))
print(np.sum(a))
print(np.min(a))
print(np.max(a))
```

```python
# [1,2,3,4,5,6] reshaped to 2x3 matrix
arr=a.reshape((2,3))
print(arr)

print(arr.flatten())  # Flatten the 2D array back to 1D
```

```python
#stack and filtter example
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c= np.array([7, 8, 9])
print(np.vstack((a, b, c)))  # Vertical stack
print(np.hstack((a, b, c)))  # Horizontal stack
arr=np.array([10, 15, 20, 25, 30, 35])
print(arr[arr>20])  # Filter elements greater than 20
```

```python
#view vs copy
original = np.array([1, 2, 3, 4, 5])
view_arr = original.view()
copy_arr = original.copy()
original[0] = 99
print("Original:", original)    # [99  2  3  4  5]
print("View:", view_arr)        # [99  2  3  4  5]
print("Copy:", copy_arr)        # [1  2  3  4  5]
```
